### Разделим футболистов по возрастным группам

```python
def age_category(age):
    if age < 20:
        return 0
    if age >= 20 and age < 30:
        return 1
    if age >= 30 and age < 36:
        return 2
    if age >= 36:
        return 3
```

```python
data['Age_category'] = data.apply(lambda x: age_category(x['Age']), axis = 1)
```

```python
# разбивка футболистов по категориям 
data['Age_category'].value_counts()
```

## Возрастные категории футболистов
Больше всего футболистов в возрасте от 20 до 30 лет,
На втором месте по количеству футболистов группа в возрасте от 30 до 36 лет, на третьем месте группа младше 20 лет, и замыкает эту классификацию группа футболистов в возрасте старше 36 лет, таких всего 23 человека.
